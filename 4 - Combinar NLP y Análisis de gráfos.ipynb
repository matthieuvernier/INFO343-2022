## 1. NLP: extraer entidades

```python
import pandas

archivo = "contaminaciónacústica_chile_2018-01-01_2021-12-31.csv"

dataset = pandas.read_csv(archivo)
dataset.drop('Unnamed: 0', axis=1, inplace=True)
dataset
```

```python
import spacy
nlp = spacy.load("es_core_news_md")
```

```python
def in_dictlist(key, value, my_dictlist):
    for entry in my_dictlist:
        if entry[key] == value:
            return entry
    return None
```

```python
persons=[]
#person= {name:"x", articles:[1,4,7]}

i=0
while i < len(dataset):
    
    text=dataset["text"][i]
    
    doc=nlp(text)
    
    #extraer entidades de tipo PERSONA
    for ent in doc.ents:
        if(ent.label_ == "PER"):
            if (" " in ent.text):
                print(ent.text)
                
                a_person = in_dictlist("name",ent.text,persons)
                
                if a_person is None:
                    persons.append({"name":ent.text,"articles":[i]})
                else:
                    if i not in a_person["articles"]:
                        a_person["articles"].append(i)

    nlp(text)
    i+=1
```

```python
persons
```

```python
for person in list(persons):
    if len(person["articles"]) <= 1:
        persons.remove(person)
```

## 2. Crear el gráfo

```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
# Añadir los nodos
for person in persons:
    G.add_node(person["name"])
```

```python
i = 0
while i < len(dataset):
    persons_in_article_i=[]
    
    for person in persons:
        if i in person["articles"]:
            persons_in_article_i.append(person["name"])
    
    for person_1 in persons_in_article_i:
        for person_2 in persons_in_article_i:
            
            #Añadir una arista
            
            if (person_1 != person_2):
                G.add_edge(person_1, person_2)
            
    i+=1
```
